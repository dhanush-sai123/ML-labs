```python
# Import Libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Load Dataset directly online
df = sns.load_dataset("titanic")

# Seaborn uses lowercase column names ('survived', 'pclass', 'sex', 'age', 'fare')
# Map them back to capitalized names to keep the rest of your code unchanged:
df = df.rename(columns={
    'survived': 'Survived',
    'pclass': 'Pclass',
    'sex': 'Sex',
    'age': 'Age',
    'fare': 'Fare'
})

# -------------------------------
# 1. Display First and Last Rows
# -------------------------------
print("First 5 Rows:")
print(df.head())

print("\nLast 5 Rows:")
print(df.tail())

# -------------------------------
# 2. Dataset Information
# -------------------------------
print("\nDataset Information:")
print(df.info())

print("\nDataset Shape:")
print(df.shape)

print("\nColumn Names:")
print(df.columns)

# -------------------------------
# 3. Missing Values
# -------------------------------
print("\nMissing Values:")
print(df.isnull().sum())

# -------------------------------
# 4. Descriptive Statistics
# -------------------------------
print("\nDescriptive Statistics:")
print(df.describe())

print("\nCategorical Statistics:")
print(df.describe(include='object'))

# -------------------------------
# 5. Duplicate Records
# -------------------------------
print("\nDuplicate Rows:", df.duplicated().sum())

# -------------------------------
# 6. Correlation Matrix
# -------------------------------
numeric_df = df.select_dtypes(include=np.number)

plt.figure(figsize=(8, 6))
sns.heatmap(numeric_df.corr(), annot=True, cmap="coolwarm")
plt.title("Correlation Matrix-24eu02025")
plt.show()

# -------------------------------
# 7. Survival Count
# -------------------------------
plt.figure(figsize=(5, 4))
sns.countplot(x='Survived', data=df)
plt.title("Survival Count-24eu02025")
plt.show()

# -------------------------------
# 8. Gender Distribution
# -------------------------------
plt.figure(figsize=(5, 4))
sns.countplot(x='Sex', data=df)
plt.title("Gender Distribution-24eu02025")
plt.show()

# -------------------------------
# 9. Passenger Class Distribution
# -------------------------------
plt.figure(figsize=(5, 4))
sns.countplot(x='Pclass', data=df)
plt.title("Passenger Class Distribution-24eu02025")
plt.show()

# -------------------------------
# 10. Age Distribution
# -------------------------------
plt.figure(figsize=(7, 5))
sns.histplot(df['Age'], bins=30, kde=True)
plt.title("Age Distribution-24eu02025")
plt.xlabel("Age")
plt.show()

# -------------------------------
# 11. Fare Distribution
# -------------------------------
plt.figure(figsize=(7, 5))
sns.histplot(df['Fare'], bins=30, kde=True)
plt.title("Fare Distribution-24eu02025")
plt.xlabel("Fare")
plt.show()

# -------------------------------
# 12. Survival by Gender
# -------------------------------
plt.figure(figsize=(6, 4))
sns.countplot(x='Sex', hue='Survived', data=df)
plt.title("Survival by Gender-24eu02025")
plt.show()

# -------------------------------
# 13. Survival by Passenger Class
# -------------------------------
plt.figure(figsize=(6, 4))
sns.countplot(x='Pclass', hue='Survived', data=df)
plt.title("Survival by Passenger Class-24eu02025")
plt.show()

# -------------------------------
# 14. Boxplot of Age
# -------------------------------
plt.figure(figsize=(6, 4))
sns.boxplot(x=df['Age'])
plt.title("Boxplot of Age-24eu02025")
plt.show()

# -------------------------------
# 15. Pairplot
# -------------------------------
pairplot_data = df[['Survived', 'Pclass', 'Age', 'Fare']].dropna()
g = sns.pairplot(pairplot_data, hue='Survived')
g.fig.suptitle("Pairplot-24eu02025", y=1.02)
plt.show()

# -------------------------------
# 16. Conclusion
# -------------------------------
print("\nEDA Completed Successfully!")
print("Observations:")
print("1. Checked missing values.")
print("2. Calculated statistical measures.")
print("3. Visualized Age, Fare, Gender, Passenger Class.")
print("4. Analyzed Survival based on Gender and Passenger Class.")
print("5. Displayed Correlation Matrix and Pairplot.")
```
